```python
# DataFrame 'df_terjemahan'

print("\nMemulai Pra-pemrosesan Teks Lanjutan (Tokenisasi, Stop Words, Stemming) pada Data Terjemahan...")

# Persiapan untuk Stop Words dan Stemming (ID)
# Mendapatkan daftar stop words Bahasa Indonesia dari NLTK
# Gunakan set untuk pencarian yang lebih cepat
stop_words_indonesian = set(stopwords.words('indonesian'))

# Membuat objek Stemmer dari Sastrawi
factory_stemmer = StemmerFactory()
stemmer = factory_stemmer.create_stemmer()

# Fungsi Pra-pemrosesan Teks Lengkap (Tokenisasi, Filter, Stop Words, Stemming)

def preprocess_text_indonesian(text):
    # Pastikan input valid
    if pd.isna(text) or not isinstance(text, str) or text.strip() == "" or text.lower() == "nan":
        return "" # Kembalikan string kosong

    # Langsung ubah ke huruf kecil dan lakukan tokenisasi
    tokens = word_tokenize(text.lower())

    # Pembersihan Token, Filter Non-Alfabet, dan Penghapusan Stop Words
    cleaned_tokens = [
        word for word in tokens
        if word.isalpha() and word not in stop_words_indonesian # Hanya kata (huruf semua) & bukan stop word
    ]

    # Stemming
    stemmed_tokens = [
        stemmer.stem(word) for word in cleaned_tokens
    ]

    # Gabungkan kembali token yang sudah diproses menjadi string per dokumen untuk Vektorisasi TF-IDF
    return " ".join(stemmed_tokens)


# Menerapkan fungsi pra-pemrosesan ke kolom terjemahan
# Pakai nama kolom yang benar dari DataFrame df_terjemahan
translated_context_col = 'translated_context' # nama kolom head
translated_response_col = 'translated_response' # sama seeperti diatas

print(f"Menerapkan pra-pemrosesan teks pada kolom '{translated_context_col}' dan '{translated_response_col}' di df_terjemahan...")

# Terapkan fungsi pada DataFrame df_terjemahan
df_terjemahan['Context_preprocessed'] = df_terjemahan[translated_context_col].apply(preprocess_text_indonesian)
df_terjemahan['Response_preprocessed'] = df_terjemahan[translated_response_col].apply(preprocess_text_indonesian)


print("\nData setelah Pra-pemrosesan Teks Lanjutan pada Data Terjemahan (contoh 5 baris pertama):")
# Tampilkan kolom terjemahan asli dan kolom baru hasil pra-pemrosesan
print(df_terjemahan[[translated_context_col, 'Context_preprocessed', translated_response_col, 'Response_preprocessed']].head())

print("\nInformasi Dataset setelah Pra-pemrosesan:")
df_terjemahan.info()

print("\nProses Pra-pemrosesan Teks Lanjutan pada Data Terjemahan selesai.")
```

```python
# Ada DataFrame 'df_terjemahan' dengan kolom 'Context_preprocessed'
# Diatas udah ada

print("\nMemulai Vektorisasi Data Context (Pra-proses) menggunakan TF-IDF...")

# Inisialisasi TfidfVectorizer
# Menggunakan parameter yang umum sebagai titik awal
# max_features: Membatasi jumlah kata (fitur) terpopuler yang akan digunakan
# min_df: Mengabaikan kata yang muncul di kurang dari 'min_df' dokumen
# max_df: Mengabaikan kata yang muncul di lebih dari 'max_df' proporsi dokumen
tfidf_vectorizer = TfidfVectorizer(max_features=5000, min_df=5, max_df=0.8)

# Melatih (fit) vectorizer pada kolom Context yang sudah dipra-proses
# dan sekaligus mengubah (transform) teks menjadi matriks TF-IDF
# Hasilnya disimpan dalam variabel tfidf_matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(df_terjemahan['Context_preprocessed'])

print(f"\nProses Vektorisasi selesai.")
print(f"Bentuk (Shape) dari matriks TF-IDF Context: {tfidf_matrix.shape}")
print(f"Jumlah fitur (kata unik) yang digunakan: {len(tfidf_vectorizer.get_feature_names_out())}")

# Menampilkan beberapa contoh kata (fitur) yang dipelajari oleh vectorizer
# Ini adalah kata-kata dasar Bahasa Indonesia dari data Context yang ada
print(f"Contoh beberapa fitur (kata-kata) yang dipelajari:")
# Ambil beberapa fitur dari tengah untuk melihat ragam kata
start_index = min(100, len(tfidf_vectorizer.get_feature_names_out()) - 50)
if start_index < 0: start_index = 0
print(tfidf_vectorizer.get_feature_names_out()[start_index : start_index + 50])
```

```python
# df_terjemahan (DataFrame dengan kolom 'Context_preprocessed' dan 'Response_preprocessed')
# tfidf_vectorizer (Objek TfidfVectorizer yang sudah di-fit pada Context_preprocessed)
# tfidf_matrix (Hasil tfidf_vectorizer.transform() pada Context_preprocessed)

print("\nImplementasi Model Chatbot (Retrieval-based) & Pencarian Kemiripan...")

# Fungsi untuk mendapatkan respons dari chatbot berdasarkan query pengguna
# Fungsi ini menerima query teks, DataFrame, vectorizer, dan matriks TF-IDF
def get_chatbot_response_with_fallback(user_query, df, tfidf_vectorizer, tfidf_matrix):
    # Pra-proses query pengguna
    preprocessed_query = preprocess_text_indonesian(user_query)
    print(f"Query pengguna setelah pra-pemrosesan: '{preprocessed_query}'") # Baris debug

    # Jika query setelah pra-proses menjadi kosong, tidak ada yang bisa dicari
    if not preprocessed_query:
        return "Maaf, saya tidak mengerti pertanyaan Anda." # Respons default jika query kosong

    # Vektorisasi query pengguna
    # Gunakan objek vectorizer yang sudah (fitted) pada data pelatihan
    # method .transform() mengubah teks baru berdasarkan kosakata dan IDF yang sudah dipelajari
    # Transform membutuhkan input dalam bentuk list/iterable, jadi bungkus preprocessed_query
    try:
        query_vector = tfidf_vectorizer.transform([preprocessed_query])
    except Exception as e:
        print(f"Error saat vektorisasi query: {e}")
        return "Maaf, terjadi kesalahan saat memproses pertanyaan Anda."


    # Hitung kemiripan antara query vector dan SEMUA vector di tfidf_matrix
    # cosine_similarity(A, B) menghitung kemiripan antara setiap baris di A dan setiap baris di B
    # Karena query_vector hanya punya 1 baris, hasilnya adalah matriks 1xN (N = jumlah baris di tfidf_matrix)
    try:
        similarity_scores = cosine_similarity(query_vector, tfidf_matrix)
    except Exception as e:
         print(f"Error saat menghitung kemiripan: {e}")
         return "Maaf, terjadi kesalahan saat mencari jawaban."


    # Temukan INDEKS (lokasi baris) dari Context di dataset pelatihan dengan skor kemiripan tertinggi
    # similarity_scores[0] mengambil array skor kemiripan untuk query kita
    # np.argmax() mencari indeks dari nilai terbesar dalam array tersebut
    most_similar_context_index = np.argmax(similarity_scores[0])

    # Ambil Response yang terkait dengan Context paling mirip dari DataFrame df_terjemahan
    # Gunakan indeks yang ditemukan untuk mengakses baris yang sesuai
    # Pastikan nama kolom Response yang diterjemahkan adalah 'translated_response' atau sesuai namamu
    translated_response_col = 'translated_response' # cek diatas

    try:
        chatbot_response = df.iloc[most_similar_context_index][translated_response_col]
    except KeyError:
        return f"Error: Kolom '{translated_response_col}' tidak ditemukan di DataFrame. Pastikan namanya benar."
    except Exception as e:
         print(f"Error saat mengambil respons: {e}")
         return "Maaf, tidak dapat mengambil jawaban yang sesuai."


    return chatbot_response

# Uji Coba Chatbot dengan beberapa Query
print("\nUji Coba Chatbot")

# Contoh query dari pengguna
query1 = "Saya merasa sangat sedih."
query2 = "Butuh bantuan untuk mengatasi kecemasan."
query3 = "Bagaimana Seseorang memulai proses konseling?"

print(f"Query Pengguna 1: {query1}")
response1 = get_chatbot_response_with_fallback(query1, df_terjemahan, tfidf_vectorizer, tfidf_matrix)
print(f"Respons Chatbot 1: {response1}\n")

print(f"Query Pengguna 2: {query2}")
response2 = get_chatbot_response_with_fallback(query2, df_terjemahan, tfidf_vectorizer, tfidf_matrix)
print(f"Respons Chatbot 2: {response2}\n")

print(f"Query Pengguna 3: {query3}")
response3 = get_chatbot_response_with_fallback(query3, df_terjemahan, tfidf_vectorizer, tfidf_matrix)
print(f"Respons Chatbot 3: {response3}\n")


print("\nUji Coba Chatbot selesai")

# Variabel 'get_chatbot_response_with_fallback' <<< merupakan fungsi yang bisa dipanggil
# dengan query pengguna untuk mendapatkan respons dari chatbot.
```

```python
# objek diatas (yang column)
# df_terjemahan (DataFrame yang sudah dimuat dan dipra-proses)
# tfidf_vectorizer (Objek TfidfVectorizer yang sudah di-fit)
# tfidf_matrix (Hasil tfidf_vectorizer.transform)
# get_chatbot_response_with_fallback (Fungsi chatbot yang sudah didefinisikan)
# preprocess_text_indonesian (Fungsi pra-pemrosesan teks)

print("\nMemodifikasi Fungsi Chatbot dengan Fallback Sederhana...")

# Menentukan Ambang Batas Kemiripan
# bisa disesuaikan (tuning) berdasarkan hasil uji coba
SIMILARITY_THRESHOLD = 0.2 # Bisa ubah nilai 0.2 ini. naik atau turun buat liat efeknya

# Fungsi Chatbot dengan Fallback Sederhana
# Menerima query teks, DataFrame, vectorizer, dan matriks TF-IDF
def get_chatbot_response_with_fallback(user_query, df, tfidf_vectorizer, tfidf_matrix, similarity_threshold=0.2):
    # Pra-proses query pengguna
    preprocessed_query = preprocess_text_indonesian(user_query)
    print(f"Query pengguna setelah pra-pemrosesan: '{preprocessed_query}'") # Baris debug

    # Jika query setelah pra-proses menjadi kosong, tidak ada yang bisa dicari
    if not preprocessed_query:
        # Fallback jika query kosong setelah pra-proses
        return "Maaf, pertanyaan Anda tidak mengandung kata kunci yang saya pahami. Bisakah Anda coba dengan kata-kata lain?"

    # Vektorisasi query pengguna
    try:
        query_vector = tfidf_vectorizer.transform([preprocessed_query])
    except Exception as e:
        print(f"Error saat vektorisasi query: {e}")
        return "Maaf, terjadi kesalahan saat memproses pertanyaan Anda."

    # Hitung kemiripan
    try:
        similarity_scores = cosine_similarity(query_vector, tfidf_matrix)
    except Exception as e:
         print(f"Error saat menghitung kemiripan: {e}")
         return "Maaf, terjadi kesalahan saat mencari jawaban."

    # Temukan Skor kemiripan paling tinggi
    max_similarity_score = np.max(similarity_scores[0])

    # Logika Fallback sederhana
    # Jika skor kemiripan tertinggi di bawah ambang batas
    if max_similarity_score < similarity_threshold:
        # Respons standar (fallback)
        print(f"Skor kemiripan tertinggi ({max_similarity_score:.4f}) di bawah ambang batas ({similarity_threshold}). Memberikan respons fallback.") # Baris debug
        return "Maaf, saya kurang paham dengan pertanyaan Anda. Bisakah Anda coba jelaskan dengan kata-kata lain atau tanyakan hal lain?"
    # Akhir LOGIKA FALLBACK

    else:
        # Jika skor kemiripan cukup tinggi, lanjutkan dengan proses retrieval normal
        print(f"Skor kemiripan tertinggi: {max_similarity_score:.4f}. Melanjutkan dengan retrieval.") # Baris debug

        # Temukan INDEKS dari Context dengan skor kemiripan tertinggi
        most_similar_context_index = np.argmax(similarity_scores[0])

        # Ambil Response yang terkait dengan Context paling mirip dari DataFrame
        translated_response_col = 'translated_response' # << sesuai kolom

        try:
            chatbot_response = df.iloc[most_similar_context_index][translated_response_col]
        except KeyError:
            return f"Error: Kolom '{translated_response_col}' tidak ditemukan di DataFrame. Pastikan namanya benar."
        except Exception as e:
             print(f"Error saat mengambil respons: {e}")
             return "Maaf, tidak dapat mengambil jawaban yang sesuai."

        return chatbot_response

# Uji Coba Chatbot yang Dimodifikasi (setelah fungsi didefinisikan)
print("\nUji Coba Chatbot dengan Fallback")

# Contoh query dari pengguna
query_topik_dataset = "Saya merasa kesepian dan sedih." # Query yang seharusnya mirip dengan dataset
query_luar_topik = "Siapa presiden pertama indonesia?" # Query yang SANGAT di luar topik dataset
query_kosong = "" # Query kosong
query_non_teks = 123 # Query bukan teks

print(f"Query Pengguna (Topik Dataset): '{query_topik_dataset}'")
# Panggil fungsi yang baru dengan query uji dan ambang batas
response_topik = get_chatbot_response_with_fallback(query_topik_dataset, df_terjemahan, tfidf_vectorizer, tfidf_matrix, SIMILARITY_THRESHOLD)
print(f"Respons Chatbot: {response_topik}\n")

print(f"Query Pengguna (Luar Topik): '{query_luar_topik}'")
response_luar = get_chatbot_response_with_fallback(query_luar_topik, df_terjemahan, tfidf_vectorizer, tfidf_matrix, SIMILARITY_THRESHOLD)
print(f"Respons Chatbot: {response_luar}\n")

print(f"Query Pengguna (Kosong): '{query_kosong}'")
response_kosong = get_chatbot_response_with_fallback(query_kosong, df_terjemahan, tfidf_vectorizer, tfidf_matrix, SIMILARITY_THRESHOLD)
print(f"Respons Chatbot: {response_kosong}\n")

print(f"Query Pengguna (Bukan Teks): {query_non_teks}")
response_non_teks = get_chatbot_response_with_fallback(query_non_teks, df_terjemahan, tfidf_vectorizer, tfidf_matrix, SIMILARITY_THRESHOLD)
print(f"Respons Chatbot: {response_non_teks}\n")


print("\nUji Coba Chatbot dengan Fallback selesai")

# Fungsi 'get_chatbot_response_with_fallback' versi yang ditingkatkan
# dengan penanganan fallback sederhana.
```
