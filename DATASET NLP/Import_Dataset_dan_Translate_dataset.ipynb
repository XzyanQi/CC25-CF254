```python
#Import Library
import pandas as pd
from googletrans import Translator
```

# Translate Dataset

```python
# Load the CSV file
file_path = '/content/train.csv'
df = pd.read_csv(file_path)

# Cek beberapa baris pertama untuk melihat struktur data
df.head()

# Inisialisasi translator
translator = Translator()

# Menerjemahkan kolom 'context' dan 'response' ke bahasa Indonesia
df['translated_context'] = df['Context'].apply(lambda x: translator.translate(x, src='en', dest='id').text)
df['translated_response'] = df['Response'].apply(lambda x: translator.translate(x, src='en', dest='id').text)

# Menampilkan beberapa hasil terjemahan
df[['Context', 'translated_context', 'Response', 'translated_response']].head()

# Jika ingin menyimpan file terjemahan baru
df.to_csv('/content/translated_train.csv', index=False)
```
